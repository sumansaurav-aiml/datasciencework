```
import re, string

def clean_str(string):
  """
  String cleaning before vectorization
  """
  try:    
    string = re.sub(r'^https?:\/\/<>.*[\r\n]*', '', string, flags=re.MULTILINE)
    string = re.sub(r"[^A-Za-z]", " ", string)         
    words = string.strip().lower().split()    
    words = [w for w in words if len(w)>=1]
    return " ".join(words)	
  except:
    return ""
```

```
#Cleaning the text
df['clean_text'] = df['text'].apply(clean_str)
df.head()
```

```
import nltk
nltk.download('stopwords')
```

```
#removing the stop words
from nltk.corpus import stopwords
stop = stopwords.words('english')

df['clean_text'] = df["clean_text"].apply(lambda words: ' '.join(word.lower() for word in words.split() if word not in stop))
```

```
df.head(5)
```

```
# Loop through all the rows and create the data frame out of the table.
data = []
for index in range(df.shape[0]):
  labels_val = [df['gender'][index], str(df['age'][index]), df['topic'][index], df['sign'][index]]
  text_val = df['clean_text'][index]
  data.append({'labels': labels_val, 'text': text_val})
```

```
final_df = pd.DataFrame(data)
```

```
final_df.head()
```

```
X = final_df.text
y = final_df.labels
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=.75,random_state=1)
```

```
y_test[0:1]
```

```
# import and instantiate CountVectorizer (with the default parameters)
from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer(ngram_range= (1, 2), stop_words= 'english', min_df = 2)
```

```
#Feed data to CountVectorizer
X_train = cvect.fit_transform(X_train)
```

```
from sklearn.preprocessing import MultiLabelBinarizer
```

```
mlb = MultiLabelBinarizer()
y_train_mlb = mlb.fit_transform(y_train)
```

```
mlb.classes_
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
lr = LogisticRegression(solver='lbfgs',max_iter=1000)
lr = OneVsRestClassifier(lr)
```

```
lr.fit(X_train, y_train_mlb)
```

```
y_test_mlb = mlb.transform(y_test)
```

```
predictions = lr.predict(X_test.toarray())
```

```
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import average_precision_score
from sklearn.metrics import recall_score
```

```
accuracy_score(y_test_mlb, predictions)
```

```
f1_score(y_test_mlb, predictions, average=None)
```

```
average_precision_score(y_test_mlb, predictions, average=None)
```

```
recall_score(y_test_mlb, predictions, average=None)
```

```
def GetPredictionAndTrue(param):
  predlabel = lr.predict(X_test[param].toarray())
  truelabel = y_test_mlb[param]
  pred=[]
  true=[]
  index = 0
  for val in predlabel[0]:
      if val == 1:
          pred.append(mlb.classes_[index])
      index = index + 1
  index = 0
  for val in truelabel:
      if val == 1:
          true.append(mlb.classes_[index])
      index = index + 1
  print('Predicted Label:',pred,'    True Label:',true)  
```
