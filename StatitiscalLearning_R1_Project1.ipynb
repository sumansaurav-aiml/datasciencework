```python
data = pd.read_csv('Laliga.csv',header=1) #used header attribute to set the second row as header
```

# Question 4

```python
def Goal_diff_count(GoalsFor , GoalsAgainst):
    return GoalsFor - GoalsAgainst
```

```python
data['GoalDiff'] = Goal_diff_count(data.GoalsFor.astype('int32'), data.GoalsAgainst.astype('int32'))
```

```python
data.sort_values('GoalDiff',ascending=False).iloc[:1,:] #this team has maximum goal diff
```

```python
data.sort_values('GoalDiff',ascending=False).iloc[-1:,:] #this team has minimum goal diff
```

# Question 5

```python
data['Winning Percent'] =  (data.GamesWon.astype('int32') / data.GamesPlayed.astype('int32'))*100
```

```python
data['Winning Percent'] = data['Winning Percent'].fillna(0)
```

```python
data.sort_values('Winning Percent',ascending=False).iloc[0:5,:]
```
